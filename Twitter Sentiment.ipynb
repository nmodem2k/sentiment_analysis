```python
import re
```

```python
  def preprocessor(text):
    text = re.sub('<[^>]*>', '', text)
    emoticons = re.findall('(?::|;|=)(?:-)?(?:\)|\(|D|P)', text)
    text = re.sub('[\W]+', ' ', text.lower()) +\
        ' '.join(emoticons).replace('-', '')
    return text
```

```python
df['SentimentText']=df['SentimentText'].apply(preprocessor)
```

```python
df=df.drop('ItemID',axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(strip_accents=None,lowercase=False,preprocessor=None,tokenizer=tokinizer_porter,use_idf=True,norm='l2',smooth_idf=True)
```

```python
y=df.Sentiment.values
```

```python
y
```

```python
X=tfidf.fit_transform(df.SentimentText)
```

```python
X[0][0:30]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.3)
```

```python
import pickle
```

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
clf=LogisticRegressionCV(cv=5,scoring='accuracy',n_jobs=-1,random_state=0,verbose=3,max_iter=300).fit(X_train,Y_train)
```

```python
saved_model=open('savedmodel.sav','wb')
```

```python
pickle.dump(clf,saved_model)
```

```python
clf.score(X_test,Y_test)
```

```python
clf.score(X_train,Y_train)
```

```python
Y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion=confusion_matrix(Y_test,Y_pred)
```
